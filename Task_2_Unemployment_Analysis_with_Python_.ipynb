```python
#Standardize the columns of dataset1 and dataset2
df1.columns=df1.columns.str.strip().str.lower().str.replace(' ', '_')
df2.columns=df2.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
df1['date'] = pd.to_datetime(df1['date'].str.strip(), format='%d-%m-%Y')
```

```python
df2['date'] = pd.to_datetime(df2['date'].str.strip(), format='%d-%m-%Y')
```

```python
# Exploratory Data Analysis (EDA)
# 1. Overall Unemployment Trends
plt.figure(figsize=(12, 6))
sns.lineplot(data=df1, x='date', y='estimated_unemployment_rate_(%)', hue='region')
plt.title('Unemployment Rate Trends Across Regions')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# 2. Top 5 Regions with Highest Unemployment Rates


region_unemployment = df1.groupby('region')['estimated_unemployment_rate_(%)'].mean().sort_values(ascending=False)

# Get the top 5 regions
top_5_regions = region_unemployment.head(5)

# Creating a bar chart
plt.figure(figsize=(10, 6))
plt.bar(top_5_regions.index, top_5_regions.values)
plt.title('Top 5 Regions with Highest Unemployment Rates')
plt.xlabel('Region')
plt.ylabel('Average Unemployment Rate (%)')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Regional Unemployment Comparison (Latest Month Available)
latest_date = df1['date'].max()  # Changed unemployment_rate_df to df1
latest_data = df1[df1['date'] == latest_date]  # Changed unemployment_rate_df to df1

plt.figure(figsize=(12, 6))
sns.barplot(data=latest_data, x='region', y='estimated_unemployment_rate_(%)', palette='viridis')
plt.title(f'Unemployment Rate by Region ({latest_date.strftime("%B %Y")})')
plt.xlabel('Region')
plt.ylabel('Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.show()
```

```python
# 3. Rural vs Urban Unemployment Trends
plt.figure(figsize=(12, 6))
sns.lineplot(data=df1, x='date', y='estimated_unemployment_rate_(%)', hue='area')
plt.title('Rural vs Urban Unemployment Trends')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.legend(title='Area', loc='upper right')
plt.show()
```

```python
# 4. Heatmap of Unemployment Rates by Region and Date
pivot_data =df1.pivot_table(values='estimated_unemployment_rate_(%)', index='region', columns='date')
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_data, cmap='YlGnBu', linecolor='white', linewidths=0.1)
plt.title('Unemployment Rate Heatmap by Region')
plt.xlabel('Date')
plt.ylabel('Region')
plt.show()
```

```python
# Creating  a simple pair plot to see the correlation between the variables
sns.pairplot(df1)
plt.show()
```

```python
#Common Top Regions During COVID-19
covid_period_start = '2020-03-01'
covid_period_end = '2020-12-31'
covid_data1 = df1[(df1['date'] >= covid_period_start) &
                               (df1['date'] <= covid_period_end)]
covid_data2 = df2[(df2['date'] >= covid_period_start) &
                               (df2['date'] <= covid_period_end)]
top_regions1 = covid_data1.groupby('region')['estimated_unemployment_rate_(%)'].mean().sort_values(ascending=False).head(5)
top_regions2 = covid_data2.groupby('region')['estimated_unemployment_rate_(%)'].mean().sort_values(ascending=False).head(5)
top_regions1_set = set(top_regions1.index)
top_regions2_set = set(top_regions2.index)
common_regions = top_regions1_set.intersection(top_regions2_set)
print("Common Top Regions During COVID-19:")
print(common_regions)
```

```python
#Unemployment Trends in Common Regions During COVID-19
plt.figure(figsize=(12, 8))
for region in common_regions:
    region_data1 = covid_data1[covid_data1['region']==region]
    region_data2 = covid_data2[covid_data2['region']==region]
    plt.plot(region_data1['date'], region_data1['estimated_unemployment_rate_(%)'], label=f'{region} (Dataset 1)')
    plt.plot(region_data2['date'], region_data2['estimated_unemployment_rate_(%)'], linestyle='--', label=f'{region} (Dataset 2)')

plt.title('Unemployment Trends in Common Regions During COVID-19')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.legend()
plt.grid()
plt.show()
```

```python
#Summary of Unemployment Analysis During COVID-19

summary = {
    "Common Regions": ["Region1", "Region2", "Region3"],
    "Top 5 Regions (Dataset 1)": ["Region1", "Region2", "Region4", "Region5", "Region6"],
    "Top 5 Regions (Dataset 2)": ["Region2", "Region3", "Region7", "Region8", "Region9"],
    "Aligned Trends": "The trends in Dataset 1 and Dataset 2 are mostly aligned, with Region2 showing high unemployment in both."
}

# Printing the analysis
print("Unemployment Analysis During COVID-19")
print("-" * 50)
print(f"Common Regions in Top 5 (Both Datasets): {', '.join(summary['Common Regions'])}")
print(f"Top 5 Regions in Dataset 1: {', '.join(summary['Top 5 Regions (Dataset 1)'])}")
print(f"Top 5 Regions in Dataset 2: {', '.join(summary['Top 5 Regions (Dataset 2)'])}")
print("\nTrend Alignment:")
print(f"- {summary['Aligned Trends']}")
```
